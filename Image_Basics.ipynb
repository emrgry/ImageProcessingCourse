Have you heard ndarrays before? Let's get dive into it.
We can create images using ndarrays. 

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

black_image = np.zeros([100,100,3], dtype=np.uint8) # this will create 100x100 (width x height) black image.
                                    # it is black because (all pixels have 0 value)
frame = cv2.cvtColor(black_image,cv2.COLOR_RGB2BGR)
plt.figure()
plt.imshow(frame)
plt.show()

#cv2.imshow("black_image", black_image)

#cv2.waitKey() 

#cv2.destroyAllWindows()
```

how can we convert this black image to the white?
if we changed pixel values to 255, image will become white.

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

#let's create a black image first and fill it with 255

white_image = np.zeros([100,100,3], dtype = np.uint8) 
white_image[:][:] = (255,255,255)

frame = cv2.cvtColor(white_image,cv2.COLOR_RGB2BGR)
plt.figure()
plt.imshow(frame)
plt.show()

#cv2.imshow("white_image", white_image)

#cv2.waitKey() 

#cv2.destroyAllWindows()
```

now we can do it for each channels which are R(red), G(green), B(blue).
_Attention_ When we processing images, image's channels arranged as BGR instead of RGB

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

red_image = np.zeros([100,100,3], dtype = np.uint8)
green_image = np.zeros([100,100,3], dtype = np.uint8)
blue_image = np.zeros([100,100,3], dtype = np.uint8)
    
# how can we convert this black image to the white?
# if we changed pixel values to 255, image will become white.
# (Blue, Green, Red)
red_image[:][:] = (0,0,255)
green_image[:][:] = (0,255,0)
blue_image[:][:] = (255,0,0)

frame1 = cv2.cvtColor(red_image,cv2.COLOR_RGB2BGR)
frame2 = cv2.cvtColor(green_image,cv2.COLOR_RGB2BGR)
frame3 = cv2.cvtColor(blue_image,cv2.COLOR_RGB2BGR)

plt.figure()
plt.imshow(frame1)
plt.figure()
plt.imshow(frame2)
plt.figure()
plt.imshow(frame3)
plt.show()

#cv2.imshow("red_image", red_image)
#cv2.imshow("green_image", green_image)
#cv2.imshow("blue_image", blue_image)

#cv2.waitKey() 

#cv2.destroyAllWindows()
```
